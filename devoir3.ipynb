Faire un graphique avec les cellules qui ont été nettoyées avec OpenRefine de la colonne "disposition" pour savoir quelles demandes d'accès à l'information ont été acceptées ou refusées. Afficher les pourcentages pour avoir les portions plus précises. 

```python
dem.disposition.value_counts().plot(kind="pie",title="Réponses des demandes d'accès à l'information")
```

```python
dem.disposition.value_counts()/len(dem.disposition)*100
```

On comprend donc qu'une majorité (56%) des informations des demandes d'accès ont été partiellement divulguées. Près d'un quart(21%) des informations ont été complètement divulguées. Une mince partie des demandes ont été complètement refusées (3%), par contre la raison que le document n'existe pas est souvent évoquée (15%). Les autres éléments illisibles dans ce graphique constituent des catégories mineures ; information non trouvée, abandon de la demande, non consultable, blanc, en progrès, traitement impossible, traitement informel et transféré. La catégorie "Autres" était des erreurs de transcription donc des informations par exemple des infos dans la mauvaise colonne.
